## Load  Data

```python
import pandas as pd 
import folium
for_map = pd.read_csv('data_map.csv', sep=',')

for_map.head()
```

```python
a= for_map['latitude']
b =for_map['longitude']
c =for_map['log_price']
zipped = zip(a,b,c)
data= list(zipped)
```

## Heatmap

```python
import folium
from folium.plugins import HeatMap

max_amount = float(for_map['log_price'].max())

hmap = folium.Map(location=[40.7, -73.9], zoom_start=9, )

hm_wide = HeatMap( data[0:25000], 
                   min_opacity=0.2,
                   max_val=max_amount,
                   radius=7.5, blur=13, 
                   max_zoom=0.5, )

#folium.GeoJson(district23).add_to(hmap)
hmap.add_child(hm_wide)
```

```python
file_path = r"geo.html"
hmap.save(file_path)     # 保存為html文件
```
